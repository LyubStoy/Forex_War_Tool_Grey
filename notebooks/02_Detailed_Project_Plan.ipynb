## Part IV — Feature Engineering

## 5. Feature Engineering

## 5.1 Returns

- Daily and 5-day EUR/USD returns.

$$
r_t = \ln \left( \frac{P_t}{P_{t-1}} \right)
$$

$$
r_{t,5} = \ln \left( \frac{P_t}{P_{t-5}} \right)
$$

## 5.2 First Differences

- Delta Spread.

$$
\Delta Spread_t = Spread_t - Spread_{t-1}
$$

## 5.3 Lagged Variables

- 1-day, 5-day, 10-day lags.

$$
X_{t-k}, \quad k \in \{1,5,10\}
$$

## 5.4 Rolling Statistics

- Moving averages.
- Rolling volatility.

$$
MA_n(t) = \frac{1}{n} \sum_{i=0}^{n-1} P_{t-i}
$$

$$
\sigma_n(t) = \sqrt{\frac{1}{n} \sum_{i=0}^{n-1} \left(r_{t-i} - \bar{r}\right)^2}
$$

## 5.5 Standardization

- Z-scores.

$$
Z_t = \frac{X_t - \mu_X}{\sigma_X}
$$

## 5.6 Momentum Indicators

- RSI and moving average signals.

$$
RSI = 100 - \frac{100}{1 + RS}
$$

## 5.7 Event Dummies

- FOMC, ECB meetings.

$$
D_t =
\begin{cases}
1, & \text{if event occurs at time } t \\
0, & \text{otherwise}
\end{cases}
$$

$$
