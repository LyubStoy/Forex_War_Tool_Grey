## Part I --- Introduction and Research Motivation

# 1. Introduction

## 1.1 Background

The foreign exchange (Forex or FX) market is the largest and most liquid financial market in the world, with daily trading volumes exceeding several trillion dollars. Unlike traditional equity markets, the FX market operates continuously across global financial centers, linking economies, financial institutions, corporations, governments, and investors through the exchange of currencies. Exchange rates play a central role in international trade, capital flows, monetary policy transmission, and global financial stability. As a result, fluctuations in currency values affect not only financial markets but also economic growth, inflation dynamics, competitiveness, and investment decisions worldwide.

Among all currency pairs, the EUR/USD exchange rate occupies a particularly important position. Representing the exchange rate between the euro and the US dollar, EUR/USD is the most actively traded currency pair globally and accounts for a substantial proportion of total FX transactions. Its importance stems from the economic weight of the Eurozone and the United States, which together represent two of the world’s largest economic blocs. Consequently, EUR/USD is often considered a benchmark indicator for global macroeconomic expectations, investor sentiment, and international capital allocation.

Movements in the EUR/USD exchange rate are influenced by a wide range of economic and financial factors. These include differences in interest rates, inflation expectations, economic growth prospects, labor market conditions, geopolitical risks, and central bank policies. Exchange rate fluctuations can have significant consequences for multinational corporations, exporters, importers, institutional investors, and policymakers. For example, a stronger euro may reduce export competitiveness for European firms, while a stronger US dollar may tighten global financial conditions due to the dominant role of the dollar in international trade and debt markets.

Because of its economic significance and the large amount of available financial and macroeconomic data, EUR/USD represents an ideal candidate for quantitative analysis and Machine Learning research. Understanding the drivers of EUR/USD dynamics may therefore provide valuable insights not only from an academic perspective, but also from a practical trading and investment standpoint.

## 1.2 Motivation

Traditional financial theory suggests that exchange rates should reflect underlying macroeconomic fundamentals. In particular, differences in interest rates, inflation, monetary policy, and economic growth between two economies are expected to influence capital flows and currency valuations over time. For this reason, macroeconomic variables are often viewed as essential explanatory factors in the study of foreign exchange markets.

One of the most important mechanisms influencing currency movements is the interest rate differential between countries. According to international finance theory, higher interest rates tend to attract foreign capital seeking better returns, thereby increasing demand for the corresponding currency. In the context of EUR/USD, differences between US Treasury yields and German or Eurozone government bond yields may therefore contain valuable information regarding future exchange rate movements. More importantly, financial markets are forward-looking: investors react not only to current interest rates, but also to expectations regarding future monetary policy decisions by the Federal Reserve (Fed) and the European Central Bank (ECB).

In recent years, global macroeconomic conditions have become increasingly complex due to events such as the COVID-19 pandemic, inflation shocks, aggressive central bank tightening cycles, geopolitical tensions, and changes in global trade dynamics. These events have significantly altered the relationships between macroeconomic indicators and financial markets, creating periods of high volatility and structural regime changes. Such environments provide an interesting opportunity to investigate whether Machine Learning techniques can capture nonlinear relationships and evolving market dynamics more effectively than traditional linear approaches.

The motivation behind this project is therefore both academic and practical. Academically, the project aims to explore the extent to which macroeconomic variables can explain and predict EUR/USD movements. Practically, the project seeks to evaluate whether these predictions can be transformed into exploitable trading signals and integrated into a realistic trading framework. Rather than developing a purely theoretical model, the objective is to approach the project from a professional and applied perspective, similar to the workflow used in quantitative finance and macro-driven trading environments.

## 1.3 Project Objectives

The primary objective of this project is to build a professional and comprehensive Machine Learning framework capable of analyzing, forecasting, and evaluating EUR/USD exchange rate movements using macroeconomic and financial variables.

The project aims to construct a complete quantitative research pipeline, including:

$$
\begin{aligned}
&\bullet \ \text{Data collection and preprocessing} \\
&\bullet \ \text{Feature engineering} \\
&\bullet \ \text{Exploratory data analysis} \\
&\bullet \ \text{Statistical testing} \\
&\bullet \ \text{Predictive modeling} \\
&\bullet \ \text{Model evaluation} \\
&\bullet \ \text{Trading signal generation} \\
&\bullet \ \text{Strategy backtesting}
\end{aligned}
$$

From a predictive perspective, the project will focus on two main tasks. The first task consists of forecasting EUR/USD returns through regression-based models. The second task involves predicting the direction of market movements (upward or downward) using classification approaches. Multiple Machine Learning techniques may be explored, including linear regression models, regularized models, tree-based methods, ensemble algorithms, and potentially more advanced approaches depending on project progression.

Another important objective is to bridge the gap between predictive accuracy and trading profitability. In financial markets, a statistically significant model does not necessarily translate into a profitable trading strategy. Therefore, the project will not only evaluate forecasting performance using statistical metrics, but will also assess the economic value of predictions through simulated trading strategies and backtesting procedures.

More broadly, the project seeks to develop a realistic ``Forex War Tool'' --- a macro-driven analytical framework designed to better understand market regimes, detect macroeconomic signals, and evaluate whether quantitative methods can provide a sustainable informational advantage in the EUR/USD market.

## 1.4 Research Questions

This project is structured around several central research questions.

The first question concerns the explanatory power of macroeconomic variables:

$$
\textit{Can macroeconomic and financial variables meaningfully explain EUR/USD movements?}
$$

While economic theory suggests that exchange rates should react to macroeconomic fundamentals, empirical evidence often shows that currency markets are noisy, nonlinear, and influenced by multiple interacting factors. The project therefore aims to investigate the strength and stability of these relationships across different market regimes.

The second question focuses on feature importance:

$$
\textit{Which macroeconomic variables contribute the most to forecasting performance?}
$$

Potential explanatory variables include interest rate spreads, inflation indicators, monetary policy variables, labor market statistics, volatility indices, and broader risk sentiment indicators. Identifying the most relevant variables may help improve both interpretability and model efficiency.

The third question addresses the practical applicability of Machine Learning predictions:

$$
\textit{Can predictive models be transformed into profitable and realistic trading strategies?}
$$

Even if a model demonstrates predictive capability, trading profitability may still be limited by market noise, transaction costs, regime shifts, and overfitting risks. The project therefore aims to evaluate whether predictive signals remain economically valuable once integrated into a simulated trading environment.

Finally, the project also seeks to explore a broader methodological question:

$$
\textit{Can Machine Learning methods provide additional value compared to traditional macroeconomic analysis in the context of foreign exchange markets?}
$$

This question is particularly important given the increasing use of data-driven techniques in modern quantitative finance.

## 1.5 Scope and Limitations

This project is intentionally ambitious and exploratory in nature. The long-term vision is to develop a sophisticated macroeconomic and Machine Learning framework capable of integrating multiple data sources, predictive models, and trading methodologies. However, due to the complexity and scale of the project, the initial implementation will focus on a limited number of core variables and research components.

The project will begin with a relatively small but economically meaningful set of explanatory variables, particularly variables related to interest rate differentials and monetary policy. This approach allows for a rigorous implementation of each stage of the Machine Learning workflow before progressively extending the framework to additional macroeconomic indicators.

It is important to emphasize that this project is not designed as a simplified academic exercise. Instead, the objective is to approach the problem from a professional quantitative finance perspective, attempting to replicate real-world research and trading workflows as closely as possible. As a result, the project may involve advanced statistical methods, extensive data preprocessing, feature engineering challenges, and significant computational experimentation.

At the same time, several limitations must be acknowledged. Financial markets are inherently noisy, adaptive, and partially unpredictable. Relationships between macroeconomic variables and exchange rates may vary over time due to structural breaks, changing market expectations, geopolitical events, and central bank interventions. Moreover, Machine Learning models are exposed to risks such as overfitting, data leakage, and instability across market regimes.

Finally, given the ambitious scope of the project and the large volume of potential research directions, there is no guarantee that every planned component will be fully implemented within the project timeline. Some models, variables, or trading frameworks may remain partially explored or reserved for future extensions. Nevertheless, the project’s primary goal remains to build a robust and realistic foundation for a professional macro-driven Machine Learning research system focused on the EUR/USD market.
