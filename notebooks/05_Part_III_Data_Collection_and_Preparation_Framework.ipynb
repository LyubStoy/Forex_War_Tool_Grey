## Part III — Data Collection and Preparation

## 3. Data Collection

The objective of this part is to define precisely which data will be collected, why these data are economically relevant, how they will be transformed, and how they will be prepared before being used in the machine learning models.

In a macro-driven EUR/USD forecasting project, data collection is not only a technical step. It is also an economic selection process. The variables must be chosen because they represent mechanisms that can influence the exchange rate: growth expectations, inflation dynamics, monetary policy expectations, yield differentials, financial stress, risk appetite, and capital flows.

The project therefore adopts a structured approach: first identifying the target variable, then selecting explanatory variables, then defining data sources, frequency treatment, and sample period.

---

## 3.1 Target Variable

The primary target variable of the project is the **EUR/USD spot exchange rate**.

The EUR/USD exchange rate represents the price of one euro expressed in US dollars. For example, if EUR/USD equals 1.0800, this means that one euro is worth 1.08 US dollars.

In this project, the EUR/USD exchange rate is not only studied as a price level. It can also be transformed into several target variables depending on the objective of the model.

The project may study:

- the EUR/USD exchange rate level;
- EUR/USD returns;
- EUR/USD directional movements;
- short-term and medium-term fluctuations.

The exchange rate level is useful for visual analysis and for understanding broad historical movements. However, for machine learning and statistical modeling, returns are often more appropriate because they measure changes rather than absolute levels.

A return measures how much the exchange rate changes from one period to another. This is important because financial markets are usually modeled through variations, not only through prices.

For example, if EUR/USD moves from 1.0800 to 1.0900, the model is interested in the change between these two values. This change may be expressed as a percentage return, a logarithmic return, or a directional movement.

Potential target transformations include:

- logarithmic returns;
- percentage returns;
- rolling returns;
- volatility-adjusted returns;
- binary classification targets.

### Logarithmic returns

Logarithmic returns are commonly used in financial analysis because they are mathematically convenient and additive over time.

$$
r_t = \ln\left(\frac{P_t}{P_{t-1}}\right)
$$

where:

- $P_t$ is the EUR/USD exchange rate at time $t$;
- $P_{t-1}$ is the EUR/USD exchange rate at the previous period;
- $r_t$ is the logarithmic return.

This transformation allows the project to focus on daily movements rather than on the exchange rate level itself.

### Percentage returns

Percentage returns measure the relative change in the exchange rate:

$$
R_t = \frac{P_t - P_{t-1}}{P_{t-1}}
$$

This can also be expressed in percentage form:

$$
R_t = \left(\frac{P_t - P_{t-1}}{P_{t-1}}\right) \times 100
$$

Percentage returns are intuitive because they directly show how much EUR/USD increased or decreased in percentage terms.

### Rolling returns

Rolling returns may be used to capture short-term or medium-term movements. For example, a 5-day rolling return measures the cumulative movement over one trading week.

$$
R_{t,5} = \frac{P_t - P_{t-5}}{P_{t-5}}
$$

This type of target can be useful if the project aims to forecast broader movements rather than only one-day changes.

### Directional target

The project may also transform EUR/USD movements into a classification target.

For example:

$$
Y_t =
\begin{cases}
1 & \text{if EUR/USD increases} \\
0 & \text{if EUR/USD decreases}
\end{cases}
$$

This allows the project to test classification models that predict whether EUR/USD is more likely to go up or down.

The objective is therefore to model and forecast EUR/USD dynamics using macroeconomic and financial explanatory variables.

## 3.2 Relative Macroeconomic Explanatory Variables

Following the theoretical framework developed in Part II, the project places primary emphasis on **relative macroeconomic variables** rather than isolated indicators.

This is a key idea in foreign exchange analysis.

EUR/USD does not depend only on whether the US economy is strong or weak. It depends on whether the US economy is stronger or weaker relative to the Euro Area.

The same logic applies to inflation, interest rates, monetary policy, employment, and market expectations.

Foreign exchange markets are comparative by nature. A currency pair always compares two economies. Therefore, EUR/USD should be analyzed through relative indicators such as:

- US growth versus Euro Area growth;
- US inflation versus Euro Area inflation;
- Fed policy versus ECB policy;
- US yields versus German yields;
- US labor market strength versus Euro Area labor market strength.

The central idea is that EUR/USD reacts mainly to **relative economic strength** and **relative monetary policy expectations**.

If US macroeconomic data are stronger than Euro Area data, this may support the US dollar and put downward pressure on EUR/USD. Conversely, if Euro Area data improve relative to US data, this may support the euro and push EUR/USD higher.

Consequently, the project will attempt to construct macroeconomic differentials between the United States and the Euro Area.

A generic differential can be written as:

$$
Differential_t = US\ Indicator_t - Euro\ Area\ Indicator_t
$$

For example:

$$
Growth\ Differential_t = US\ GDP\ Growth_t - Euro\ Area\ GDP\ Growth_t
$$

or:

$$
Inflation\ Differential_t = US\ CPI_t - Euro\ Area\ HICP_t
$$

These differentials will allow the project to study whether relative macroeconomic conditions help explain EUR/USD movements.

## 3.2.1 Relative Growth Indicators

Growth indicators measure the strength of economic activity. They are important because stronger growth can influence exchange rates through several channels.

First, stronger growth may attract capital inflows. Investors may prefer to invest in an economy with better growth prospects.

Second, stronger growth may increase expectations of tighter monetary policy. If an economy grows strongly, its central bank may be more likely to raise interest rates or keep rates high.

Third, stronger growth can improve confidence in the currency.

For EUR/USD, the project will focus on comparable indicators between the United States and the Euro Area.

Potential variables include:

### GDP Growth Differential

$$
GDP\ Growth\ Differential_t = US\ GDP\ Growth_t - Euro\ Area\ GDP\ Growth_t
$$

GDP growth is one of the broadest indicators of economic performance. If US GDP growth is stronger than Euro Area GDP growth, this may support the dollar. If Euro Area GDP growth is stronger, this may support the euro.

However, GDP is released quarterly, so it is not very frequent. It may be more useful for medium-term analysis than for short-term daily forecasting.

### Manufacturing Activity Differential

Manufacturing PMIs provide timely information about industrial activity.

$$
Manufacturing\ Differential_t = US\ Manufacturing\ PMI_t - Euro\ Area\ Manufacturing\ PMI_t
$$

A PMI above 50 usually indicates expansion, while a PMI below 50 suggests contraction.

If the US Manufacturing PMI improves relative to the Euro Area PMI, this may indicate stronger US industrial momentum.

### Services Activity Differential

Services activity is especially important because modern economies are heavily service-based.

$$
Services\ Differential_t = US\ Services\ PMI_t - Euro\ Area\ Services\ PMI_t
$$

A stronger US services sector relative to the Euro Area may support the dollar, while stronger Euro Area services activity may support the euro.

### Business Activity Differential

The project may also compare US ISM indicators with European business indicators.

The ISM Manufacturing and ISM Services indices are closely watched by markets. For Europe, comparable indicators may include business confidence indices, composite PMIs, or European Commission sentiment indicators.

### Consumption Differential

Retail sales measure household consumption.

$$
Consumption\ Differential_t = US\ Retail\ Sales_t - Euro\ Area\ Retail\ Sales_t
$$

Consumption is important because it represents a major component of GDP. Stronger retail sales may indicate stronger domestic demand.

### Industrial Production Differential

Industrial production measures output in manufacturing, mining, and utilities.

$$
Industrial\ Production\ Differential_t = US\ Industrial\ Production_t - Euro\ Area\ Industrial\ Production_t
$$

This indicator may be particularly relevant during periods of energy shocks, supply chain disruption, or industrial slowdown.

### Consumer Confidence Differential

Consumer confidence reflects households’ expectations about income, employment, and the economy.

A higher US consumer confidence level relative to the Euro Area may indicate stronger consumption prospects in the United States.

### Labor Market Differential

The labor market is a central variable for monetary policy.

For the United States, Nonfarm Payrolls are one of the most important monthly indicators. For the Euro Area, employment growth and unemployment statistics may be used.

A stronger US labor market relative to the Euro Area may increase expectations of tighter Fed policy, which can support the dollar.

### Unemployment Differential

$$
Unemployment\ Differential_t = US\ Unemployment\ Rate_t - Euro\ Area\ Unemployment\ Rate_t
$$

This variable must be interpreted carefully. A higher unemployment rate is usually negative for a currency because it signals economic weakness.

Therefore, the sign of the variable matters. In some cases, it may be better to use:

$$
Labor\ Strength\ Differential_t = Euro\ Area\ Unemployment_t - US\ Unemployment_t
$$

so that a higher value indicates stronger US labor market conditions.

## 3.2.2 Relative Inflation Indicators

Inflation is one of the most important drivers of exchange rates because it directly influences central bank decisions.

If inflation is higher in the United States than in the Euro Area, markets may expect the Federal Reserve to maintain higher interest rates. This can support the dollar.

If inflation is higher in the Euro Area than in the United States, markets may expect the ECB to remain more restrictive, which can support the euro.

The project will therefore analyze several inflation differentials.

### Inflation Differential

$$
Inflation\ Differential_t = US\ CPI_t - Euro\ Area\ HICP_t
$$

The US CPI and Euro Area HICP are the main consumer inflation indicators.

This differential captures whether inflationary pressure is stronger in the United States or in the Euro Area.

### Core Inflation Differential

Core inflation excludes volatile components such as food and energy.

$$
Core\ Inflation\ Differential_t = US\ Core\ CPI_t - Euro\ Area\ Core\ HICP_t
$$

Core inflation is especially important for central banks because it reflects more persistent inflation pressure.

### Monetary Inflation Differential

The project may also include US Core PCE inflation, which is closely followed by the Federal Reserve.

The Fed often gives strong importance to PCE inflation because it is considered a broad measure of consumer price dynamics.

### Wage Inflation Differential

Wage growth is important because it can generate persistent inflation.

If wages grow faster in the United States than in the Euro Area, this may create stronger inflation pressure in the US economy and influence Fed expectations.

### Inflation Expectations Differential

Inflation expectations are important because markets react not only to current inflation but also to expected future inflation.

If US inflation expectations rise relative to Euro Area expectations, this may affect bond yields and EUR/USD.

### Energy Inflation Sensitivity

Energy prices can affect the United States and the Euro Area differently.

The Euro Area is often more sensitive to imported energy prices, especially during geopolitical or supply shocks. Therefore, oil and gas price movements may have asymmetric effects on EUR/USD.

## 3.2.3 Relative Monetary Policy Indicators

Monetary policy is one of the most direct channels influencing exchange rates.

Currencies are strongly affected by expected interest rate differentials. If US interest rates are expected to be higher than Euro Area rates, dollar-denominated assets may become more attractive. This can support the dollar and push EUR/USD lower.

The project will study several monetary policy indicators.

### Policy Rate Differential

$$
Policy\ Rate\ Differential_t = Fed\ Funds\ Rate_t - ECB\ Deposit\ Rate_t
$$

This variable captures the difference between the main policy rates of the Federal Reserve and the European Central Bank.

A higher differential in favor of the United States may support the dollar.

### 2-Year Yield Differential

$$
2Y\ Yield\ Differential_t = US\ 2Y\ Yield_t - German\ 2Y\ Yield_t
$$

The 2-year yield differential is particularly important because short-term yields are strongly influenced by monetary policy expectations.

This variable is one of the core signals of the project.

If the US 2-year yield rises relative to the German 2-year yield, markets may interpret this as stronger Fed tightening expectations relative to the ECB. This can support the dollar and put downward pressure on EUR/USD.

### 10-Year Yield Differential

$$
10Y\ Yield\ Differential_t = US\ 10Y\ Yield_t - German\ 10Y\ Yield_t
$$

The 10-year yield differential captures longer-term expectations, including growth, inflation, risk premia, and long-term monetary policy expectations.

It may be more useful for medium-term EUR/USD trends.

### Forward Guidance Differential

Central banks influence markets not only through actual rate decisions but also through communication.

Forward guidance refers to the way central banks signal future policy intentions.

The project may attempt to quantify the relative tone of Fed and ECB communication, although this is more complex and may require text analysis or event-based coding.

### OIS Expectations Differential

Overnight Index Swaps reflect market expectations of future central bank rates.

An OIS differential can be useful because it captures what markets expect the Fed and ECB to do in the future.

### Real Yield Differential

Real yields adjust nominal yields for inflation expectations.

$$
Real\ Yield = Nominal\ Yield - Expected\ Inflation
$$

The real yield differential may be very important because investors care about inflation-adjusted returns.

## 3.2.4 Additional Financial and Risk Variables

Besides macroeconomic fundamentals, EUR/USD is also affected by global risk sentiment and financial conditions.

During risk-off periods, investors may move toward safe-haven assets. The US dollar often benefits from global risk aversion.

Therefore, the project may include additional market variables.

### VIX

The VIX measures expected volatility in the US equity market.

A rising VIX usually indicates higher market stress. In many periods, higher risk aversion supports the US dollar.

### Financial Stress Indices

Financial stress indices summarize tensions in financial markets, including credit spreads, volatility, liquidity stress, and banking sector risks.

These indices can help identify periods when EUR/USD is driven more by risk sentiment than by traditional macroeconomic variables.

### Credit Spreads

Credit spreads measure the difference between risky corporate bond yields and safer government bond yields.

Widening credit spreads usually indicate rising financial stress.

### Gold Prices

Gold may be used as a safe-haven indicator.

Gold prices can also reflect real yield expectations and dollar movements.

### Oil Prices

Oil prices affect inflation, trade balances, and growth expectations.

They may influence the Euro Area more strongly during energy shocks because of its dependence on energy imports.

### CDS Spreads

Credit Default Swap spreads measure perceived default risk.

Sovereign CDS spreads may help capture country-specific stress, especially in Europe.

### Geopolitical Risk Indices

Geopolitical shocks can influence EUR/USD through energy prices, safe-haven flows, and uncertainty.

### Economic Policy Uncertainty Indices

Policy uncertainty can reduce investment, increase risk aversion, and affect currency markets.

## 3.2.5 Capital Flow Variables

Exchange rates are also influenced by international capital flows.

Even if macroeconomic indicators are important, actual flows of money between regions can strongly affect currency demand.

This part is optional because capital flow data can be more difficult to collect, less frequent, and sometimes harder to align with daily market data.

Potential variables include:

### Treasury International Capital Flows

TIC data measure foreign investment flows into US securities.

Strong foreign demand for US assets can support the dollar.

### Foreign Purchases of US Treasury Bonds

Foreign purchases of US Treasuries may increase demand for dollars.

### ETF Fund Flows

ETF flows may capture investor allocation toward US or European assets.

### Cross-Border Portfolio Flows

These flows show how international investors allocate capital across regions.

### Current Account Differentials

The current account reflects trade balances, income flows, and transfers.

A stronger current account position may support a currency over the long term.

## 3.2.6 Research Scope and Progressive Expansion

The project aims to integrate a wide range of macroeconomic and financial variables.

However, the number of potential explanatory variables is very large. It would not be realistic to implement all indicators immediately with the same level of quality.

Therefore, the project adopts a progressive research methodology.

The first step is to begin with the most economically meaningful variables. These include:

- EUR/USD exchange rate;
- US 2-year yield;
- German 2-year yield;
- US–Germany 2-year yield spread;
- inflation indicators;
- growth indicators;
- monetary policy indicators.

Once the core framework is built, the project can progressively add more variables.

The objective is not to include as many variables as possible without structure. The objective is to construct a coherent macroeconomic forecasting framework.

This means that each variable should be justified economically before being included in the model.

The project will therefore follow this logic:

1. Start with a limited but meaningful set of variables.
2. Clean and align the data properly.
3. Test the economic relationship with EUR/USD.
4. Add new variables progressively.
5. Compare model performance.
6. Avoid unnecessary complexity.

This progressive approach is important because adding too many variables too early may create noise, overfitting, missing data problems, and interpretation difficulties.

## 3.4 Data Frequency

The project will primarily rely on daily financial market data.

However, macroeconomic variables are released at different frequencies.

Some variables are daily, such as:

- EUR/USD exchange rate;
- government bond yields;
- VIX;
- oil prices;
- gold prices.

Some variables are weekly, such as:

- initial jobless claims;
- certain financial flow indicators.

Some variables are monthly, such as:

- CPI inflation;
- unemployment rate;
- retail sales;
- industrial production;
- PMI indicators.

Some variables are quarterly, such as:

- GDP growth;
- current account data.

This creates an important methodological problem: the project must combine variables with different frequencies into one consistent dataset.

Several techniques may be used.

### Forward filling

Forward filling means that the latest available macroeconomic value is carried forward until a new value is released.

This reflects the idea that markets only know the latest published value.

### Release-date synchronization

The project should avoid using data before it was actually released.

For example, if March inflation is published in April, the model should not use the March value before its publication date.

This is important to avoid look-ahead bias.

### Event alignment

Some macroeconomic indicators may be treated as event variables.

For example, Nonfarm Payrolls can be aligned with the exact release date to study the market reaction around the announcement.

### Market reaction windows

The project may analyze how EUR/USD reacts around macroeconomic releases.

For example:

- same-day reaction;
- 1-day reaction;
- 5-day reaction;
- 10-day reaction.

This can help identify whether some variables have immediate or delayed effects.

## 4. Data Cleaning and Alignment

The objective of this section is to transform raw macroeconomic and financial data into a coherent, structured, and exploitable dataset suitable for econometric and machine learning analysis.

Because the project combines multiple financial and macroeconomic datasets originating from different institutions and market providers, several preprocessing operations are required before modeling can begin.

These operations include:

- date parsing;
- missing value treatment;
- type conversion;
- dataset merging and alignment;
- construction of the final master dataset.

Data preparation represents one of the most important stages of the project because the quality of the final dataset directly influences the reliability of the statistical and machine learning models.

---

## 4.1 Date Parsing

Macroeconomic and financial datasets frequently use heterogeneous date formats depending on the data provider.

Examples may include:

- standard calendar dates;
- monthly dates;
- quarterly periods;
- timestamps including hours and minutes.

Before merging datasets, all dates must therefore be converted into a standardized datetime format.

The objectives of date parsing are:

- ensuring chronological consistency;
- enabling time-series operations;
- facilitating dataset merging;
- preventing parsing and synchronization errors.

Once standardized, dates are typically:

- sorted chronologically;
- indexed as time-series observations;
- aligned with trading days or release dates.

Proper date parsing is essential because time-series analysis entirely depends on correct chronological ordering.

Incorrect date management may generate severe distortions in forecasting models and statistical analysis.

---

## 4.2 Missing Values, Frequency Alignment and Resampling

Macroeconomic and financial datasets frequently contain missing observations.

Missing values may arise because:

- variables possess different frequencies;
- financial markets close during holidays;
- historical coverage differs across providers;
- some indicators begin later than others;
- publication schedules differ between economies.

Missing observations can create major difficulties because many machine learning algorithms cannot process incomplete data directly.

Several treatment methods may therefore be considered.

### Forward Filling

Forward filling consists of propagating the most recent available observation forward until a new value is released.

This method is commonly used in macroeconomic analysis because markets only observe the latest published information.

### Backward Filling

Backward filling may occasionally be applied in specific situations.

However, this method must be used cautiously because it may introduce future information into historical observations.

### Interpolation

Interpolation techniques may estimate missing intermediate values for continuous variables.

These methods can smooth certain datasets but may also distort actual market behavior if applied excessively.

### Row Deletion

Observations containing missing values may also be removed entirely.

However, excessive deletion can significantly reduce the sample size and weaken statistical robustness.

The project must therefore balance:

- data completeness;
- statistical reliability;
- avoidance of look-ahead bias;
- preservation of historical information.

---

## 4.3 Type Conversion

Financial datasets frequently contain variables stored in non-numerical formats.

Typical issues include:

- percentage symbols;
- commas used as decimal separators;
- textual labels;
- string-based missing values;
- inconsistent formatting conventions.

Before statistical analysis, all variables must be converted into coherent numerical formats.

Type conversion is essential because machine learning models require numerical inputs.

Incorrect formatting may generate:

- calculation errors;
- merge inconsistencies;
- statistical distortions;
- model training failures.

Consequently, the project must systematically verify that each variable is correctly interpreted and formatted before inclusion in the final dataset.

---

## 4.4 Merge and Alignment

One of the most important stages of the project is the construction of a unified chronological dataset integrating all macroeconomic and financial variables.

This step is particularly complex because variables possess:

- heterogeneous frequencies;
- different publication schedules;
- different market calendars;
- unequal historical coverage.

The project must therefore create a coherent time-series structure where all variables are properly synchronized.

Several alignment principles are necessary.

### Trading-Day Alignment

Financial variables may be aligned according to trading days only.

This avoids distortions created by weekends and market holidays.

### Release-Date Alignment

Macroeconomic variables must only become available after their official publication date.

This is essential to avoid look-ahead bias, where the model would incorrectly access information unavailable at the forecasting date.

### Lag Construction

Certain explanatory variables may be intentionally lagged in order to simulate realistic forecasting conditions.

Lagged variables may also help capture delayed market reactions.

### Rolling Window Alignment

Rolling averages and rolling indicators may be constructed to smooth short-term volatility and better capture medium-term dynamics.

These transformations may improve signal extraction and reduce market noise.

Proper alignment is fundamental because incorrectly synchronized datasets may completely invalidate forecasting results.
