## Part II — Economic and Theoretical Framework

# 2. Economic Foundations of the EUR/USD

## 2.1 Introduction — From Yield Spreads to Relative Macroeconomic Dynamics

In a previous research project, \textbf{Forex\_Grey Project}, an empirical analysis was conducted on the relationship between the EUR/USD exchange rate and the US--Germany 2-Year yield spread.

The study demonstrated that the yield spread explains a meaningful portion of EUR/USD fluctuations and that the relationship is both economically intuitive and statistically significant.

The core result can be summarized as follows:

$$
\text{Higher US yields relative to German yields}
\rightarrow
\text{Stronger USD relative to EUR}
\rightarrow
\text{Lower EUR/USD}
$$

The analysis confirmed that the US--Germany 2-Year yield spread is one of the most important macro-financial signals driving EUR/USD dynamics.

However, the project also revealed an important limitation:

the yield spread itself is not the original source of market information.

Bond yields and yield spreads are continuously adjusted by financial markets according to expectations regarding future monetary policy decisions by the Federal Reserve and the European Central Bank.

Those monetary policy expectations are themselves driven by macroeconomic expectations regarding:

- economic growth;
- labor market conditions;
- inflation;
- consumption;
- production;
- business activity.

As a consequence, attempting to forecast EUR/USD exclusively through yield spreads may lead to delayed reactions because financial markets may have already incorporated a large portion of the information into bond prices.

Therefore, the present project adopts a more upstream macroeconomic approach.

Rather than focusing primarily on yield spreads themselves, the objective is to study the macroeconomic forces that generate monetary policy expectations before they become fully reflected in interest rates and bond markets.

## Relative Macroeconomic Logic

Foreign exchange markets are fundamentally relative markets.

The US dollar does not strengthen simply because the US economy performs well.

The dollar strengthens mainly when:

- the US economy performs better than the Euro Area economy;
- US inflation pressures exceed European inflation pressures;
- the Federal Reserve is expected to be more restrictive than the ECB;
- US yields become more attractive relative to European yields.

For this reason, the project introduces a central theoretical shift:

instead of studying isolated macroeconomic indicators independently, the project will focus on comparative US--Euro Area macroeconomic dynamics.

The objective is to construct and analyze relative macroeconomic differentials capable of explaining and potentially anticipating EUR/USD movements.

## Research Philosophy of the Project

The central philosophy of this project is the following:

$$
\textit{
The earlier the macroeconomic information enters the forecasting process,
the greater the potential informational advantage before markets fully
price the information into bond yields and exchange rates.
}
$$

Consequently, the project places strong emphasis on:

- economic activity indicators;
- comparative macroeconomic dynamics;
- macroeconomic surprise indicators;
- expectation revisions;
- relative economic momentum.

The ultimate objective is to identify which macroeconomic differentials between the United States and the Euro Area contain the strongest explanatory and predictive power for EUR/USD movements.

# 2.2 Relative Macroeconomic Determinants of EUR/USD

## 2.2.1 Relative Growth Expectations

Growth expectations are among the most important drivers of exchange rates because they directly influence:

- capital flows;
- corporate investment;
- monetary policy expectations;
- bond yields.

The project will focus primarily on highly comparable US and Euro Area growth indicators.

---

## Comparative Growth Indicators

- **GDP Growth Differential:** US GDP Growth vs Euro Area GDP Growth
- **Manufacturing PMI Differential:** US Manufacturing PMI vs Euro Area Manufacturing PMI
- **Services PMI Differential:** US Services PMI vs Euro Area Services PMI
- **Manufacturing Activity Differential:** ISM Manufacturing vs EU Manufacturing Activity
- **Services Activity Differential:** ISM Services vs EU Services Activity
- **Retail Sales Differential:** US Retail Sales vs Euro Area Retail Sales
- **Industrial Production Differential:** US Industrial Production vs Euro Area Industrial Production
- **Consumer Confidence Differential:** US Consumer Confidence vs Euro Area Consumer Confidence
- **Labor Market Differential:** US Nonfarm Payrolls vs Euro Area Employment Indicators
- **Unemployment Differential:** US Unemployment Rate vs Euro Area Unemployment Rate

---

## Relative Growth Mechanism

$$
\text{Stronger US Growth Relative to Europe}
\rightarrow
\text{Higher Expected Fed Tightening}
\rightarrow
\text{Higher US Yields Relative to Europe}
\rightarrow
\text{USD Appreciation}
\rightarrow
\text{EUR/USD Decline}
$$

---

## 2.2.2 Relative Inflation Dynamics

Inflation differentials strongly influence expectations regarding future monetary policy divergence between the Federal Reserve and the ECB.

The project will therefore compare inflationary pressures between the two economic zones.

---

## Comparative Inflation Indicators

- US CPI vs Euro Area HICP
- US Core CPI vs Euro Area Core HICP
- US Core PCE Inflation vs Euro Area Core Inflation
- US Wage Growth vs Euro Area Wage Growth
- US Inflation Expectations vs Euro Area Inflation Expectations
- Energy Inflation Sensitivity

---

## Inflation Transmission Mechanism

$$
\text{Higher US Inflation Relative to Europe}
\rightarrow
\text{More Aggressive Fed Expectations}
\rightarrow
\text{Higher US Rates Relative to ECB Rates}
\rightarrow
\text{Wider Yield Spread}
\rightarrow
\text{Stronger USD}
$$

---

## 2.2.3 Relative Monetary Policy Expectations

Monetary policy divergence is one of the direct transmission channels between macroeconomic conditions and currency valuation.

The project will therefore analyze comparative monetary policy expectations between the Federal Reserve and the ECB.

---

## Comparative Monetary Indicators

- US 2Y Treasury Yield vs German 2Y Yield
- US 10Y Treasury Yield vs German 10Y Yield
- Fed Forward Guidance vs ECB Forward Guidance
- OIS Expectations
- Real Yield Differentials

---

## 2.2.4 Risk Sentiment and Financial Conditions

Although macroeconomic activity remains the dominant structural driver, financial market conditions may temporarily dominate exchange rate movements during periods of uncertainty or crisis.

Potential variables include:

- VIX
- Financial Stress Indices
- Credit Spreads
- Safe-Haven Flows
- Gold Prices
- CDS Spreads
- Geopolitical Risk Indices

---

## 2.2.5 International Capital Flows

Exchange rates are also influenced by cross-border capital movements.

Potential variables include:

- Foreign Purchases of US Treasury Bonds
- ETF Fund Flows
- Portfolio Investment Flows
- Current Account Differentials

The project may additionally consider structural and geopolitical variables capable of temporarily distorting macroeconomic relationships:

- Banking sector stress
- Sanctions
- Supply chain disruptions
- Fiscal policy divergence

# 2.3 Strategic Research Orientation

The strategic objective of this project is therefore to move upstream in the macro-financial transmission mechanism.

Instead of relying only on yield spreads, the framework attempts to identify:

- the macroeconomic drivers behind monetary policy expectations;
- the relative economic dynamics between the United States and the Euro Area;
- the indicators capable of anticipating future market repricing.

The project will attempt to integrate and analyze as many relevant macroeconomic indicators as possible.

However, given the scale and complexity of the project, it is important to recognize that not every planned variable or modeling component may ultimately be fully implemented.

The project therefore adopts an ambitious but realistic research philosophy:

$$
\textit{
Build a professional macroeconomic and machine learning framework
while progressively expanding the number of variables,
models, and market relationships studied.
}
$$

The primary objective remains to construct a robust and coherent macro-driven forecasting and trading framework for EUR/USD rather than attempting to exhaustively model every possible market factor.
