## 🎯 Project Objective and Scope

$$
\Large \textbf{Macro-Driven Machine Learning Framework for EUR/USD Forecasting and Trading}
$$

---

## 1. Project Philosophy

The purpose of this project is to build a professional-grade Machine Learning framework designed to analyze, model, and predict the behavior of the EUR/USD exchange rate using macroeconomic and financial variables.

This project is intentionally conceived as a real-world quantitative research project, rather than as a simple academic exercise.

Its objectives are threefold:

$$
\begin{aligned}
&1.\ \textbf{Deepen understanding of the macroeconomic drivers of EUR/USD.} \\
\\
&2.\ \textbf{Apply the full Machine Learning workflow end to end.} \\
\\
&3.\ \textbf{Develop a reusable and extensible research framework capable of analyzing} \\
&\textbf{and predicting EUR/USD dynamics, which can be progressively enriched over} \\
&\textbf{time with additional variables, datasets, and models.}
\end{aligned}
$$

This project also builds upon a previous research phase focused on the relationship between the US--Germany 2-year yield spread and the EUR/USD exchange rate.

The initial study already explored the statistical relationship between yield spread dynamics and EUR/USD movements through correlation analysis, differencing techniques, rolling statistics, and regression models.

Rather than reproducing the same analysis from scratch, the present project adopts a progressive research approach by extending the framework to additional macroeconomic and financial variables in order to construct a broader and more realistic predictive system.

The previous research project can be found here:

$$
\textbf{Forex Grey Project Repository}
$$

\url{https://github.com/LyubStoy/Forex_Grey}

This approach reflects real-world quantitative research practices, where existing research components are progressively expanded, refined, and integrated into larger analytical frameworks.

## 5. Initial Scope: Limited Number of Variables

To ensure that every stage of the workflow is implemented rigorously, the project will begin with a deliberately limited set of highly relevant explanatory variables.

The initial research phase of the previous project already focused extensively on the core yield spread framework involving EUR/USD, US 2-Year Treasury yields, German 2-Year Government Bond yields, and the US--Germany 2-Year spread.

As a result, the present project will not aim to reproduce the same analysis in detail. Instead, the emphasis will progressively shift toward the integration and study of additional macroeconomic and financial variables in order to build a broader and more realistic predictive framework.

### Core Variables (Previously Studied Framework --- Baseline Reference)

$$
\begin{aligned}
&\bullet\ \text{EUR/USD exchange rate} \\
\\
&\bullet\ \text{US 2-Year Treasury Yield} \\
\\
&\bullet\ \text{German 2-Year Government Bond Yield} \\
\\
&\bullet\ \text{Yield Spread (US 2Y $-$ DE 2Y)}
\end{aligned}
$$

### Additional Variables (Primary Focus of the Present Project)

$$

&\bullet\ \text{Euro Area HICP Inflation} \\

&\bullet\ \text{ECB Deposit Rate} \\
\\
&\bullet\ \text{Nonfarm Payrolls} \\
\\

The objective is therefore to progressively extend the original yield spread framework toward a more comprehensive macro-financial Machine Learning system capable of capturing a broader range of economic dynamics influencing EUR/USD behavior.

## 6. Important Note Regarding Feature Expansion

A broad set of explanatory variables is identified from the outset.

However:

$$
\textbf{It is unlikely that all of these variables will be fully processed and incorporated into the first version of the project.}
$$

This is intentional.

The key objective is to:

$$
\begin{aligned}
&1.\ \textbf{Build a robust and reusable framework.} \\
\\
&2.\ \textbf{Traverse every major stage of the Machine Learning lifecycle.} \\
\\
&3.\ \textbf{Establish a methodology that can easily accommodate additional variables in future versions.}
\end{aligned}
$$

Once the architecture is complete, integrating new features will require only limited additional work.
